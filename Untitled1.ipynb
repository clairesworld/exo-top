```python
import sympy as sp
CMF = sp.Symbol('CMF')
M_p = sp.Symbol('M_p')
D_l0 = sp.Symbol('D_l0')
d = sp.Symbol('d')
R_E = sp.Symbol('R_E')
```

```python
d_exp = (1.07 - 0.121*CMF)*M_p**(1/3.7)*R_E - D_l0 - (1.07*CMF**0.5 - 0.121*CMF**(3/2))*M_p**(1/3.7)*R_E
CMF_exp = sp.solve(d_exp, CMF)
```
