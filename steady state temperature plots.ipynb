```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 300e3
z = np.linspace(0, d) # m
k = 3
rho_litho = 2700
af = 2e-6 # radiogenic heating at tf in W m^-3
a0 = af/(np.exp(-4.5/2.94))
print('A0=', a0)
T_sfc = 300
q0 = 10e-3 # W m^-2

T = -a0/(2*k)*z**2 + (q0+a0*d)/k*z + T_sfc

plt.plot(T, -z*1e-3)
plt.xlabel("T (K)")
plt.ylabel("z (km)")
```
