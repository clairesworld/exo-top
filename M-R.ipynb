```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
M_p = np.linspace(0.5, 10) # earth masses

# in earth radii
def R_Otengi(M):
    return 1.03*M**0.29 # for rho < 3.3 g cm^-3

def R_Seager(M):
    return(10**(-0.20945 + 1/3*np.log10(M/6.41) - 0.0804*(M/6.41)**0.394))*2.84 # for CMF=0.3 Fe core, MgSiO3 mantle

def R_Zeng(M, CMF=0.3):
    return (1.07 - 0.21*CMF)*M**(1/3.7)

print('Otengi 2019, R(M=1)=', R_Otengi(1))
print('Zeng 2016, R(M=1)=', R_Zeng(1))
print('Seager 2008, R(M=1)=', R_Seager(1))


fig, ax = plt.subplots(1,1)
ax.plot(M_p, R_Otengi(M_p), label='Otengi 2019')
ax.plot(M_p, R_Zeng(M_p), label='Zeng 2016')
ax.plot(M_p, R_Seager(M_p), label='Seager 2007')
ax.plot(1, 1, 'kv')
ax.legend(frameon=False)
ax.set_xlabel('Mass (M$_E$)')
ax.set_ylabel('Radius (R$_E$)')
```
